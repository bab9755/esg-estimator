```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tickers = ['NVDA', 'MSFT', 'GOOGL', 'AMZN', 'META', 'TSLA', 'BRK-B', 
 'XOM', 'COST', 'JNJ', 'CRM', 'CVX', 'NOW', 'AMD', 'DIS', 'ABT', 
 'ADBE', 'ISRG', 'INTU', 'DHR', 'SPGI', 'ANET', 'PGR']

df = yf.download(tickers, period="5y")
```

```python
df.head()
```

```python
print(df.columns)
close_prices = df['Close']
close_prices
```

```python
close_prices[close_prices.index >= "2020-01-01"].plot(figsize=(15,10));
```

```python
from pypfopt import risk_models
from pypfopt import plotting

sample_covariance = risk_models.sample_cov(close_prices, frequency=252)
sample_covariance
```

```python
plotting.plot_covariance(sample_covariance, plot_correlation=True)
```

```python
from pypfopt import expected_returns

mu = expected_returns.capm_return(close_prices)
mu
```

```python

from pypfopt import EfficientFrontier

S = risk_models.CovarianceShrinkage(close_prices).ledoit_wolf()
ef = EfficientFrontier(mu, S, weight_bounds = (0,1))

ef.max_sharpe()
weights = ef.clean_weights()
weights
```

```python
ef.portfolio_performance(verbose=True)
```

```python
esg_scores = {}
first_esg = 0
for ticker in tickers:
    try:
        stock = yf.Ticker(ticker)
        esg_data = stock.sustainability
        if esg_data is not None:
            print(f"type: {type(esg_data)}")
            first_esg = esg_data
            
            
            esg_scores[ticker] = esg_data.loc["totalEsg", "esgScores"]
        else:
            break
    except Exception as e:
        print(f"Error fetching ESG for {ticker}: {e}")
        break
        

esg_scores



        
```

```python
for values in esg_scores.values():
    print(values)
```

```python
esg_df = pd.DataFrame(list(esg_scores.items()), columns=["Ticker", "Total ESG Score"])
esg_df
```

```python
portfolio_min_score = 10.0
esg_score_values = esg_df['Total ESG Score'].tolist()
esg_score_values
```

```python
ef_esg = EfficientFrontier(mu, S)
ef_esg.add_constraint(lambda w: esg_score_values @ w >= portfolio_min_score)
ef_esg.max_sharpe()
```

```python
ef_esg.portfolio_performance(verbose=True)
```

```python
ef = EfficientFrontier(mu, S, weight_bounds = (0,1))
fig, ax = plt.subplots()
ef_max_sharpe = ef.deepcopy()
plotting.plot_efficient_frontier(ef, ax=ax, show_assets=True)

ef_max_sharpe.max_sharpe()
ret_tangent, std_tangent, _ = ef_max_sharpe.portfolio_performance()
ax.scatter(std_tangent, ret_tangent, marker="*", s=100, c="r", label="Max Sharpe")

ef_esg_aggressive = EfficientFrontier(mu, S)
portfolio_min_score = 10.0
ef_esg_aggressive.add_constraint(lambda w: esg_score_values @ w >= portfolio_min_score)
plotting.plot_efficient_frontier(ef_esg_aggressive, ax=ax, show_assets=False)


ef_esg_mild = EfficientFrontier(mu, S)
portfolio_mild_score = 5.0
ef_esg_mild.add_constraint(lambda w: esg_score_values @ w >= portfolio_mild_score)
plotting.plot_efficient_frontier(ef_esg_mild, ax=ax, show_assets=False)
```

```python
ef_esg_mild = EfficientFrontier(mu, S)
portfolio_min_score = 5.0
ef_esg_mild.add_constraint(lambda w: esg_score_values @ w >= portfolio_min_score)
ef_esg_mild.max_sharpe()
esg_mild = ef_esg_mild.portfolio_performance(verbose = True)
esg_mild
```

```python
ef_esg_aggressive = EfficientFrontier(mu, S)
portfolio_min_score = 20.0
ef_esg_aggressive.add_constraint(lambda w: esg_score_values @ w >= portfolio_min_score)
ef_esg_aggressive.max_sharpe()
esg_agg = ef_esg_aggressive.portfolio_performance(verbose = True)
esg_agg
```

```python
ef = EfficientFrontier(mu, S, weight_bounds = (0,1))
ef.max_sharpe()
no_esg = ef.portfolio_performance(verbose = False)
```
